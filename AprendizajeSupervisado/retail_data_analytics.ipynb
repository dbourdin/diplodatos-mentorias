## Contenido del Dataset
Se provee de datos históricos de ventas de 45 tiendas localizadas en diferentes regiones, cada tienda contiene un número determinado de departamentos. Cada compañía además organiza eventos de rebajas a lo largo del año. Estas rebajas preceden a feriados destacados como lo son: el Super Bowl, el Día del Trabajo, Día de Acción de Gracias y Navidad. Las semanas que incluyen estos feriados tienen un peso 5 veces mayor que las que no.

El contenido de los cuatro archivos (Stores, Users, Features y Sales) se presenta a continuación:

### Stores
Información anónima de las 45 tiendas, incluye tipo y tamaño de las mismas.

### Users
Lista de usuarios y las tiendas que frecuentan.

### Features
Contiene información adicional relacionada a las tiendas, la localidad y actividad regional para las fechas dadas.
- Store: el número de tienda.
- Date: la semana correspondiente.
- Temperatura: temperatura promedio de la región.
- Fuel Price: costo del combustible en la región.
- Markdown 1-5: datos anónimos relacionados a rebajas promocionales. Esta información esta sólo disponible a partir de Noviembre del 2011 y no esta disponible para todas las tiendas todo el tiempo. Lo valores faltantes están marcados como NA.
- CPI: índice de precios al consumidor.
- Unemployment: tasa de desempleo.
- IsHoliday: si la semana contiene al menos un día feriado.

### Sales
Datos historicos de ventas, cubren desde el 05-02-2010 hasta el 01-11-2012. Incluye los siguientes campos:
- Store: el número de tienda.
- Dept: el número de departamento.
- Date: la fecha correspondiente a la transacción.
- Weekly_Sales: ventas para el departamento dado en la tienda correspondiente.
- IsHoliday: indica si corresponde a una semana con días feriados o no.

Link a la fuente: https://www.kaggle.com/manjeetsingh/retaildataset#Features%20data%20set.csv

```python
import os

import matplotlib.pyplot as plt
import seaborn
import numpy as np
import pandas as pd
# import fbprophet
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import mean_squared_error, r2_score, classification_report
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn import svm
from sklearn.preprocessing import PolynomialFeatures
import graphviz
from ml.visualization import plot_confusion_matrix
from ml.visualization import corrplot

%matplotlib inline
```

Comenzamos con la importación del dataset generado a partir del trabajo anterior

```python
CURRENT_DIR = os.path.dirname('__file__')
DATASET_PATH = os.path.join(CURRENT_DIR, '../datasets')
sales_features = pd.read_csv(os.path.join(DATASET_PATH, 'sales_features.csv'), parse_dates=['Date'])
```

Creamos las columnas Year, Month y Week

```python
def set_row_year_and_week(row):
    year, week, day = row.Date.isocalendar()
    row['Year'] = year
    row['Month'] = row.Date.month
    row['Week'] = week
    return row
```

```python
#sales_features.drop_duplicates(subset=['Date', 'IsHoliday'], inplace=True)
sales_features = sales_features.apply(set_row_year_and_week, axis=1)
```

Seteamos el valor de la semilla

```python
np.random.seed(10) 
```

```python
sales_features.head(5)
```

Aplicamos coeficiente de correlación a los datos para entender si existe asociación lineal entre las variables:

```python
# Convertimos la fecha a numérica para crear la matriz de correlación
sales_features.Date = pd.to_numeric(sales_features.Date)
```

```python
corr = sales_features[['Weekly_Sales', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'IsHoliday', 'IsMarkdown', 'Date', 'Week', 'Month', 'Year', 'Store']].corr(method='spearman')
corr
```

```python
plt.figure(figsize=(12, 6))
corrplot(corr)
```

```python
# Reconvertimos a datetime
sales_features.Date = pd.to_datetime(sales_features.Date)
```

Creamos columnas de variables dummies para las variables categoricas Store, semana y mes: 

```python
sales_features['Store'] = pd.Categorical(sales_features['Store'])
store_dummies = pd.get_dummies(sales_features['Store'], prefix = 'Store')

sales_features['Month'] = pd.Categorical(sales_features['Month'])
month_dummies = pd.get_dummies(sales_features['Month'], prefix = 'Month')

sales_features['Week'] = pd.Categorical(sales_features['Week'])
week_dummies = pd.get_dummies(sales_features['Week'], prefix = 'Week')
```

Las unimos al dataset sales_features

```python
sales_features = pd.concat([sales_features, store_dummies, month_dummies, week_dummies], axis=1)
```

```python
sales_features.head(5)
```

Separamos en conjunto de entrenamiento y evaluación

```python
# División entre instancias y etiquetas
X, y = sales_features.iloc[:, 2:], sales_features[['Weekly_Sales', 'Year']]

# división entre entrenamiento y evaluación
X_train = X[X.Year.isin([2010, 2011])]
X_test = X[X.Year == 2012]
y_train = y[y.Year.isin([2010, 2011])]
y_test = y[y.Year == 2012]

# Borramos Year de los conjuntos de lo que queremos predecir
y_train = y_train['Weekly_Sales']
y_test = y_test['Weekly_Sales']

# La feature del dataset a analizar será "Week" 
#feature = 'Week'
# X_train_feature = X_train[feature].values.reshape(-1, 1)
# X_test_feature = X_test[feature].values.reshape(-1, 1)
```

```python
# Nos aseguramos que los conjuntos de test y validación tengan la misma cantidad de registros que los valores a predecir
print(f'Total de registros en X_train: {len(X_train)}')
print(f'Total de registros en y_train: {len(y_train)}')
print(f'Total de registros en X_test: {len(X_test)}')
print(f'Total de registros en y_test: {len(y_test)}')
```

Se puede observar que en general no hay fuerte correlacion entre las variables. Vamos entonces a realizar una regresión lineal para ver si encontramos relaciones mas robustas.

Creamos el listado de features que vamos a usar en el modelo

```python
features = [feature for feature in sales_features.columns if 'Week_' in feature or 'Month_' in feature or 'Store_' in feature]
features += ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'IsHoliday', 'IsMarkdown']
```

### Modelo de regresion lineal

```python
model = LinearRegression()
X_train_features = X_train[features]
X_test_features = X_test[features]
model.fit(X_train_features, y_train)
```

```python
# print(r2_score(y_train, model.predict(np.array(X_train.Temperature).reshape(-1, 1))))
# print(r2_score(y_test, model.predict(np.array(X_test.Temperature).reshape(-1, 1))))
print('Análisis de R2 score')
print('Conjunto de entrenamiento', r2_score(y_train, model.predict(X_train_features)))
print('Conjunto de validación', r2_score(y_test, model.predict(X_test_features)))
print()
print('Análisis de la media del error cuadrado')
print('Conjunto de entrenamiento', mean_squared_error(y_train, model.predict(X_train_features)))
print('Conjunto de validación', mean_squared_error(y_test, model.predict(X_test_features)))
```

El modelo de regresion lineal arroja muy buenos resultados para el conjunto de entrenamiento (R2 = 94%) pero no funciona bien para los datos de test donde el R2 es negativo y el error cuadrado medio mucho mas alto.

### Polynomial Regression

```python
polynomial_degree = 2
poly_features = PolynomialFeatures(polynomial_degree)
poly_features.fit(X_train_features)

X_poly_train = poly_features.transform(X_train_features)
X_poly_test = poly_features.transform(X_test_features)

model = LinearRegression()
model.fit(X_poly_train, y_train)
```

```python
print('Análisis de R2 score')
print('Conjunto de entrenamiento', r2_score(y_train, model.predict(X_poly_train)))
print('Conjunto de validación', r2_score(y_test, model.predict(X_poly_test)))
print()
print('Análisis de la media del error cuadrado')
print('Conjunto de entrenamiento', mean_squared_error(y_train, model.predict(X_poly_train)))
print('Conjunto de validación', mean_squared_error(y_test, model.predict(X_poly_test)))
```

Si bien obtenemos un R2 mucho mas alto que con el modelo de regresion lineal (casi cercano a uno) para el conjunto de entrenamiento, en el conjunto de validacion esta medida tiene muy mala performance.

### Arbol de regresión

```python
model = DecisionTreeRegressor()
model.fit(X_train_features, y_train)
```

```python
print('Análisis de R2 score')
print('Conjunto de entrenamiento', r2_score(y_train, model.predict(X_train_features)))
print('Conjunto de validación', r2_score(y_test, model.predict(X_test_features)))
print()
print('Análisis de la media del error cuadrado')
print('Conjunto de entrenamiento', mean_squared_error(y_train, model.predict(X_train_features)))
print('Conjunto de validación', mean_squared_error(y_test, model.predict(X_test_features)))
```

En este caso obtenemos un R2 aceptable tanto en el conjunto de datos de entrenamiento como en el de validación. Por lo tanto elegimos este modelo para encontrar sus mejores parametros.

### Busqueda en grilla de los mejores hiperparametros para arbol de regresion

```python
plt.figure(figsize=(14, 4), dpi= 80, facecolor='w', edgecolor='k')

for idx, criterion in enumerate(('mse', 'friedman_mse'), start=1):
    exploring_params = {
        'max_depth': [None, 10, 15],
        'min_samples_leaf': [1,2,3,4,5],  #
    }
    m = DecisionTreeRegressor(criterion=criterion) 
    model = GridSearchCV(m, exploring_params, cv=5, scoring='r2')
    model.fit(X_train_features, y_train)
    
    print("# Exploración de hiperparámetros para función de coste \"%s\"" % criterion, end="\n\n")
    
    print("Mejor conjunto de parámetros:")
    print(model.best_params_, end="\n\n")

    print("Puntajes de la grilla:", end="\n\n")
    means = model.cv_results_['mean_test_score']
    stds = model.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, model.cv_results_['params']):
        print("Exactitud: %0.3f (+/-%0.03f) para los parámetros %r" % (mean, std ** 2, params))
    print()

#     print("Reporte de clasificación para el mejor clasificador (sobre conjunto de evaluación):", end="\n\n")
#     y_true, y_pred = y_test, model.predict(X_test_features)
#     print(classification_report(y_true, y_pred), end="\n\n")
    
    
#     print("================================================", end="\n\n")

#     plt.subplot(1, 3, idx)
#     plot_confusion_matrix(confusion_matrix(y_true, y_pred),
#                           classes=['No Feriado', 'Feriado'], title="Matriz de confusión para %s" % criterion)
```
