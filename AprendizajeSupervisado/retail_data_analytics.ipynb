## Contenido del Dataset
Se provee de datos históricos de ventas de 45 tiendas localizadas en diferentes regiones, cada tienda contiene un número determinado de departamentos. Cada compañía además organiza eventos de rebajas a lo largo del año. Estas rebajas preceden a feriados destacados como lo son: el Super Bowl, el Día del Trabajo, Día de Acción de Gracias y Navidad. Las semanas que incluyen estos feriados tienen un peso 5 veces mayor que las que no.

El contenido de los cuatro archivos (Stores, Users, Features y Sales) se presenta a continuación:

### Stores
Información anónima de las 45 tiendas, incluye tipo y tamaño de las mismas.

### Users
Lista de usuarios y las tiendas que frecuentan.

### Features
Contiene información adicional relacionada a las tiendas, la localidad y actividad regional para las fechas dadas.
- Store: el número de tienda.
- Date: la semana correspondiente.
- Temperatura: temperatura promedio de la región.
- Fuel Price: costo del combustible en la región.
- Markdown 1-5: datos anónimos relacionados a rebajas promocionales. Esta información esta sólo disponible a partir de Noviembre del 2011 y no esta disponible para todas las tiendas todo el tiempo. Lo valores faltantes están marcados como NA.
- CPI: índice de precios al consumidor.
- Unemployment: tasa de desempleo.
- IsHoliday: si la semana contiene al menos un día feriado.

### Sales
Datos historicos de ventas, cubren desde el 05-02-2010 hasta el 01-11-2012. Incluye los siguientes campos:
- Store: el número de tienda.
- Dept: el número de departamento.
- Date: la fecha correspondiente a la transacción.
- Weekly_Sales: ventas para el departamento dado en la tienda correspondiente.
- IsHoliday: indica si corresponde a una semana con días feriados o no.

Link a la fuente: https://www.kaggle.com/manjeetsingh/retaildataset#Features%20data%20set.csv

```python
# División entre instancias y etiquetas
X, y = store_sales.iloc[:, 2:], store_sales[['Weekly_Sales', 'Year']]

# división entre entrenamiento y evaluación
X_train = X[X.Year.isin([2010, 2011])]
X_test = X[X.Year == 2012]
y_train = y[y.Year.isin([2010, 2011])]
y_test = y[y.Year == 2012]

# Borramos Year de los conjuntos de lo que queremos predecir
y_train = y_train['Weekly_Sales']
y_test = y_test['Weekly_Sales']

# La feature del dataset a analizar será "Week" 
feature = 'Week'
X_train_feature = X_train[feature].values.reshape(-1, 1)
X_test_feature = X_test[feature].values.reshape(-1, 1)
```

A esta altura, ya hemos separado nuestros conjuntos de entrenamiento y validación, nos aseguramos que la cantidad de elementos a predecir coincida con la cantidad de elementos en los conjuntos de features, y analizamos la correlación entre las variables.
Se puede observar una fuerte correlación negativa entre la temperatura y las ventas. Vamos entonces a realizar una regresión lineal utilizando esta feature.

```python
model = LinearRegression()
X_train_features = X_train[['Temperature', 'Fuel_Price', 'IsHoliday']]
X_test_features = X_test[['Temperature', 'Fuel_Price', 'IsHoliday']]
# model.fit(np.array(X_train[['Temperature', 'Fuel_Price', 'IsHoliday']]).reshape(-1, 1), y_train)
model.fit(X_train_features, y_train)
```

```python
# print(r2_score(y_train, model.predict(np.array(X_train.Temperature).reshape(-1, 1))))
# print(r2_score(y_test, model.predict(np.array(X_test.Temperature).reshape(-1, 1))))
print('Análisis de R2 score')
print('Conjunto de entrenamiento', r2_score(y_train, model.predict(X_train_features)))
print('Conjunto de validación', r2_score(y_test, model.predict(X_test_features)))
print()
print('Análisis de la media del error cuadrado')
print('Conjunto de entrenamiento', mean_squared_error(y_train, model.predict(X_train_features)))
print('Conjunto de validación', mean_squared_error(y_test, model.predict(X_test_features)))
```

Vamos a probar utilizando Polynomial Regression

```python
polynomial_degree = 5
poly_features = PolynomialFeatures(polynomial_degree)
poly_features.fit(X_train_features)

X_poly_train = poly_features.transform(X_train_features)
X_poly_test = poly_features.transform(X_test_features)

model = LinearRegression()
model.fit(X_poly_train, y_train)
```

```python
print('Análisis de R2 score')
print('Conjunto de entrenamiento', r2_score(y_train, model.predict(X_poly_train)))
print('Conjunto de validación', r2_score(y_test, model.predict(X_poly_test)))
print()
print('Análisis de la media del error cuadrado')
print('Conjunto de entrenamiento', mean_squared_error(y_train, model.predict(X_poly_train)))
print('Conjunto de validación', mean_squared_error(y_test, model.predict(X_poly_test)))
```
